```python
# =========================================================
# CONFIGURATION
# =========================================================

wallet_filter = "1 = 1"


# =========================================================
# LOAD NORMALIZED DATA
# =========================================================

spark.sql(f"""
INSERT INTO raw_data_normalized_v2

WITH source_data AS (

    SELECT

        r.company,
        r.wallet_bank,
        TO_DATE(r.date)             AS transaction_date,
        TO_DATE(r.depositdate)      AS depositdate,
        TO_DATE(r.withdrawaldate)   AS withdrawaldate,
        r.pay_id,
        r.status,

        CAST(r.amount AS DECIMAL(18,4))                    AS amount,
        r.currency,
        CAST(r.commission AS DECIMAL(18,4))               AS commission,
        r.commissioncurrency,
        CAST(r.exchangerate AS DECIMAL(18,4))             AS exchangerate,
        
        r.project,
        r.notes,
        r.reference,
        r.paymentmethod,

        CAST(r.settlement AS DECIMAL(18,4))               AS settlement,
        CAST(r.settlementcomission AS DECIMAL(18,4))      AS settlementcomission,
        CAST(r.topups AS DECIMAL(18,4))                   AS topups,
        CAST(r.topupscomission AS DECIMAL(18,4))          AS topupscomission,
        CAST(r.affilates AS DECIMAL(18,4))                AS affilates,
        CAST(r.affilatescommission AS DECIMAL(18,4))      AS affilatescommission,
        CAST(r.chargebacksrefunds AS DECIMAL(18,4))       AS chargebacksrefunds,
        CAST(r.chargebacksrefundscommission AS DECIMAL(18,4))
            AS chargebacksrefundscommission,
        CAST(r.expenses AS DECIMAL(18,4))                 AS expenses,
        CAST(r.expensescommission AS DECIMAL(18,4))       AS expensescommission,

        r.notes222,
        r.notes333,
        r.notes444,
        r.path,
        r.ingection_id,
        r.ingection_time,
        r.raw_hash

    FROM transfer_data_to_fabric r

    WHERE {wallet_filter}

)

SELECT

    company,
    wallet_bank,
    transaction_date,
    depositdate,
    withdrawaldate,
    pay_id,
    status,
    amount,
    currency,
    commission,
    commissioncurrency,
    exchangerate,

    CAST(
        ROUND(amount * exchangerate, 4)
        AS DECIMAL(18,4)
    ) AS finalamount,

    project,
    notes,
    reference,
    paymentmethod,
    settlement,
    settlementcomission,
    topups,
    topupscomission,
    affilates,
    affilatescommission,
    chargebacksrefunds,
    chargebacksrefundscommission,
    expenses,
    expensescommission,
    notes222,
    notes333,
    notes444,
    path,
    ingection_id,
    ingection_time,
    raw_hash

FROM source_data
""")
```
